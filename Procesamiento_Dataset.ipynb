```python
#extraccion de estadisticas
import pandas as pd

for col in ['about', 'goal', 'motivation', 'description']:
    df[col] = df[col].fillna('').astype(str)

def text_length_stats(df, columns, status=None):
    if status:
        df = df[df['status'] == status]
    stats = {}
    for col in columns:
        lengths = df[col].apply(len)
        stats[col] = {
            'mean': lengths.mean(),
            'min': lengths.min(),
            'max': lengths.max()
        }
    return stats

campos = ['about', 'goal', 'motivation', 'description']

stats_general = text_length_stats(df, campos)

stats_funded = text_length_stats(df, campos, status='funded')

stats_unfunded = text_length_stats(df, campos, status='unfunded')
def mostrar_stats(nombre, stats):
    print(f"\n{nombre.upper()}:")
    for campo, valores in stats.items():
        print(f"{campo}: Media={valores['mean']:.2f}, Mínimo={valores['min']}, Máximo={valores['max']}")

mostrar_stats("Media general", stats_general)
mostrar_stats("Proyectos fundados", stats_funded)
mostrar_stats("Proyectos no fundados", stats_unfunded)
```

```python
# Aplicamos limpieza a los datos brutos


import re

def quitarURL(text):
  text = re.sub(r'http\S+', '', text)


  return text
def quitarPar(text):
  text = re.sub(r'\(\s*\)', '', text)
  return text

def remove_chars(text):
    text = re.sub(r'!\[\]\(', '', str(text)) # Remove "![]("
    return text
def remove_extra_chars(text):
    text = re.sub(r'[\n\r\t]', ' ', str(text)) # Replace \n, \r, \t with space
    text = re.sub(r'\s+', ' ', text) # Replace multiple spaces with single space
    return text.strip()

def remove_asterisks(text):
    text = re.sub(r'\*', '', str(text))  # Remove asterisks
    return text

for col in ['about', 'description', 'goal', 'motivation']:
  df[col] = df[col].astype(str).apply(quitarURL)
  df[col] = df[col].astype(str).apply(quitarPar)
  df[col] = df[col].astype(str).apply(remove_chars)
  df[col] = df[col].astype(str).apply(remove_extra_chars)
  df[col] = df[col].astype(str).apply(remove_asterisks)

```

```python
# detectamos idiomas
import fasttext
import pandas as pd
model = fasttext.load_model('lid.176.ftz')

def detect_language(text):
    try:
        label, confidence = model.predict(str(text))
        return label[0].replace("__label__", "")
    except:
        return "unknown"

columns_to_detect = ['about', 'description', 'goal', 'motivation']

language_df = pd.DataFrame()

for col in columns_to_detect:
    language_df[col] = df[col].apply(detect_language)

language_df



language_df['main_language'] = language_df.mode(axis=1)[0]

df['idioma'] = language_df.mode(axis=1)[0]
language_stats = language_df['main_language'].value_counts()

print(language_stats)



```

```python
# cargamos modelos traductores


tokenizers = {
    'ca': MarianTokenizer.from_pretrained('Helsinki-NLP/opus-mt-ca-es'),
    'en': MarianTokenizer.from_pretrained('Helsinki-NLP/opus-mt-en-es'),
    'gl': MarianTokenizer.from_pretrained('Helsinki-NLP/opus-mt-gl-es'),

    'fr': MarianTokenizer.from_pretrained('Helsinki-NLP/opus-mt-fr-es')
}

models = {
    'ca': MarianMTModel.from_pretrained('Helsinki-NLP/opus-mt-ca-es'),
    'en': MarianMTModel.from_pretrained('Helsinki-NLP/opus-mt-en-es'),
    'gl': MarianMTModel.from_pretrained('Helsinki-NLP/opus-mt-gl-es'),

    'fr': MarianMTModel.from_pretrained('Helsinki-NLP/opus-mt-fr-es')
}
```

```python
# Aplicamos traduccion
def conditional_translate(row, text_column):
    lang = row['idioma']
    text = row[text_column]

    if lang in ['gl','fr'] and isinstance(text, str) and text.strip():
        tokenizer = tokenizers[lang]
        model = models[lang]
        batch = tokenizer([text], return_tensors="pt", padding=True, truncation=True)
        generated = model.generate(**batch)
        print(f'traducido ({lang})')
        return tokenizer.decode(generated[0], skip_special_tokens=True)
    else:
        print('no traducido')
        return text
columns_to_translate = ['about', 'description', 'goal', 'motivation']

for col in columns_to_translate:

    df[col] = df.apply(lambda row: conditional_translate(row, col), axis=1)
```
